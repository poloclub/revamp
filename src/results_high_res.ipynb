### This notebook plots results of texture attacks

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from matplotlib.patches import Polygon
import pandas as pd
from utils.plotting import plot_scores, plot_ngon

colors = [
"#f0f0f0"
, "#d9d9d9"
, "#bdbdbd"
, "#969696"
, "#737373"
, "#525252"
, "#252525"
, "#000000" ]
```

```python
results_range = range(5)
results_dir = "../results/person/16_seq_sensor_pos_512/"
title = "% Score of Target Class Detections - 48 Sensor Positions (seq)"
label = "Person"
seq_16_bicycle_data = plot_scores(results_range, results_dir, title, label)
```

```python
rows, columns = 1, 5
fig, axs = plt.subplots(rows, columns, figsize=(20, 4))
for i in range(rows*columns):
    ax = axs[i]
    ax.set_title("Pass {}".format(i))
    plt.suptitle("Successful Target Prediction by Sensor Position: Person Class (seq)")
    plot_ngon(num_sides=16, radius=1, results_path = f"../results/person/16_seq_sensor_pos_512/{i}_scores.csv", annotations=seq_16_bicycle_data[i], ax=ax)
```
